```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("train.csv")
df
```

```python
df['TotalBath'] = df['BsmtFullBath'] + df['BsmtHalfBath'] + df['FullBath'] + df['HalfBath']
df['TSquareFoot'] =  df['1stFlrSF'] + df['TotalBsmtSF'] + df['2ndFlrSF'] + df['GrLivArea'] 
```

```python
df['TotalBath']
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['TotalBath','TSquareFoot','BedroomAbvGr']],df.SalePrice)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
test_df =pd.read_csv("test.csv")
```

```python
test_df['TotalBath'] = test_df['BsmtFullBath'] + test_df['BsmtHalfBath'] + test_df['FullBath'] + test_df['HalfBath']
test_df['TSquareFoot'] =  test_df['1stFlrSF'] + test_df['TotalBsmtSF'] + test_df['2ndFlrSF'] + test_df['GrLivArea'] 
test_df['TotalBath'] = test_df['TotalBath'].fillna(0)
test_df['TSquareFoot'] = test_df['TSquareFoot'].fillna(0)
test_df['BedroomAbvGr'] = test_df['BedroomAbvGr'].fillna(0)
```

```python
predicted_prices = reg.predict(test_df[['TotalBath', 'TSquareFoot', 'BedroomAbvGr']])
```

```python
print(predicted_prices)
```

```python
# Save predictions to a CSV file
test_df['PredictedPrice'] = predicted_prices
test_df[['PredictedPrice']].to_csv("predicted_prices.csv", index=False)
```

```python
r2_score_training = reg.score(df[['TotalBath', 'TSquareFoot', 'BedroomAbvGr']], df['SalePrice'])
print("R² Score on Training Data:", r2_score_training)
```
